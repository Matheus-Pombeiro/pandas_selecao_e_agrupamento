## First class - knowing the data

```python
# Importing Pandas
import pandas as pd
```

```python
# Getting the dataset
emissoes_gases = pd.read_excel('/content/drive/MyDrive/Estudos/Data Science/Colab Notebooks/Pandas/Pandas: selecionando e agrupando dados/1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx', sheet_name='GEE Estados')
```

```python
# Finding the unique values in 'Nível 1 - Setor' column
emissoes_gases['Nível 1 - Setor'].unique()
```

```python
# Finding the unique values in 'Estado' column
emissoes_gases['Estado'].unique()
```

```python
# Filtering the dataset by southern Brazilian region
emissoes_gases[emissoes_gases['Estado'].isin(['PR', 'SC', 'RS'])].head(3)
```

```python
# Filtering the dataset by 'Mudança de Uso da Terra e Floresta' entries in the 'AM' state
emissoes_gases[(emissoes_gases['Nível 1 - Setor'] == 'Mudança de Uso da Terra e Floresta') & (emissoes_gases['Estado'] == 'AM')].head(3)
```

```python
# Finding the max value in 'Agropecuaria' in the 'Para' state
emissoes_gases.loc[(emissoes_gases['Nível 1 - Setor'] == 'Agropecuária') & (emissoes_gases['Estado'] == 'PA'), 2021].max()
```

## Second class - grouping data

```python
# Grouping the dataset by the 'Nível 1 - Setor' using groupby and groups
nivel_1_setor_index = emissoes_gases.groupby('Nível 1 - Setor').groups
nivel_1_setor_index
```

```python
# Grouping the dataset by 'Agropecuária' entries
agropecuaria = emissoes_gases.groupby('Nível 1 - Setor').get_group('Agropecuária')
agropecuaria.head(3)
```

```python
# Grouping the dataset by '2021' entries and calculating the mean
mean_2021 = emissoes_gases.groupby('Nível 1 - Setor')[[2021]].mean()
mean_2021.head(3)
```

```python
# Grouping the dataset by 'Nível 1 - Setor' entries, calculating the sum of each economic activity and sorting the values
economic_activities_sum = emissoes_gases.groupby('Nível 1 - Setor').sum(numeric_only=True).sort_values('Nível 1 - Setor', ascending=False)
economic_activities_sum.head(3)
```

## Third class - multi index clustering

```python
# Transforming the dataframe into a long one
colunas_info = list(emissoes_gases.loc[:, 'Nível 1 - Setor':'Produto'].columns)
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021].columns)
emissoes_por_ano = emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name='Ano', value_name='Emissão')
emissoes_por_ano.head(3)
```

```python
# Transforming the long dataframe into a wide one using 'pivot_table'
emissao_setores = emissoes_por_ano.pivot_table(values='Emissão', index='Ano', columns='Nível 1 - Setor', aggfunc='mean')
emissao_setores.head()
```

```python
# Plotting the previous dataframe
emissao_setores.plot(subplots=True, figsize=(10, 40))
```

```python
# Grouping the dataset
emissoes_estado_setor = emissoes_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
emissoes_estado_setor
```

```python
# Selecting an element inside a multi index dataframe
emissoes_estado_setor.xs('Energia', level=1)
```

```python
# Selecting the maximum value using 'idxmax'
emissoes_estado_setor.xs('MG', level=0).idxmax()
```

```python
# Finding the maximum index values in each state
emissoes_estado_setor = emissoes_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
emissoes_estado_setor.groupby(level=0).idxmax()
```

```python
# Finding the maximum index values in each economic activity
emissoes_estado_setor = emissoes_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
emissoes_estado_setor.groupby(level=1).idxmax()
```
